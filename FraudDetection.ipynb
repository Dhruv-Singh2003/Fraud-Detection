```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix, roc_auc_score, roc_curve
```

```python
# Load the dataset (update path if needed)
df = pd.read_csv('newcreditcard.csv')
```

```python
#Print 1st 5 rows of dataset
df.head()
```

```python
#Print the information of dataset
df.info()
```

```python
#Checking the null value in each column
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical Measure of the transactions
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values of both transactions
df.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#training model
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Then split the scaled data
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Training Data Accuracy:",training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score of test data:",test_data_accuracy)
```
